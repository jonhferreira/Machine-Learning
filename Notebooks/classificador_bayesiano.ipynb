```python
class classifier_bayes:
  dataset = []
  tam_dataset = 0
  labels = [] 
  n_atrib = 0

  def fit(self,dataset,labels):
    self.dataset = dataset.copy()
    self.tam_dataset = len(self.dataset)
    self.labels = labels 
    self.n_atrib = len(self.dataset[0])

  # probabilidade de P(x|Ci)
  def prob_ind(self,x,atb_valor,Ci):
    cont = 0
    for amos in self.dataset:
      if (amos[x] == atb_valor) and (self.labels[x] == Ci):
        cont += 1
        
    #errado, nao Ã© dataset,mas todas as tuplas com Ci
    return (cont/self.tam_dataset)

  
  # probabilidade condicional P(X|Ci)
  def prob_cond(self,X,Ci):
    prod = 1
    for i in range(self.n_atrib):
      prod *= self.prob_ind(i,X[i],Ci)
    
    return prod

  # teorema de bayes
  def teor_bayes(self,X,Ci,P_X):
    P_Ci = self.labels.count(Ci)/self.tam_dataset
  
    P_Ci_X = (self.prob_cond(X,Ci)*P_Ci)/P_X

    return P_Ci_X

  def predict(self, features):
    labels_test = []

    # rotulos disponiveis para a classificacao
    labels_unique = list(set(self.labels))

    for feat in features:
      P_X = self.dataset.count(feat)/self.tam_dataset
      max_prob = -float("inf")
      max_lab = ''
      for Ci in labels_unique:
        res = self.teor_bayes(feat,Ci,P_X)
        if res > max_prob:
          max_prob = res
          max_lab = Ci
      
      labels_test.append(max_lab)

    return labels_test


  

```

```python

# criacao de classificador
bayes = classifier_bayes()

feat_train = [[180,70],[185,80],[190,95],[160,45],[155,50],[150,60],[181,100],[180,80],[160,50]]
label_train = [1,1,1,2,2,2,1,1,2]
```

```python
# treinando classificador
bayes.fit(feat_train,label_train)
```

```python
# realizando teste de previsao com classificador

feat_test = [[180,80],[160,45]]
label_test = [1,2]
labels = bayes.predict(feat_test)

print("previsao: ",labels)
print("real: ",label_test)
```
